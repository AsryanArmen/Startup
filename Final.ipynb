```python
def employee_count(data):
    if pd.isnull(data['Team size all employees']) and pd.notnull(data['Employees count MoM change']):
        data['Team size all employees'] = data['Employee Count'] + int(data['Employees count MoM change'])
    if pd.isnull(data['Team size all employees']) and pd.isnull(data['Employees count MoM change']):
        data['Employees count MoM change'] = 0
        data['Team size all employees'] = data['Employee Count']
    if pd.notnull(data['Team size all employees']):
        data['Employees count MoM change'] = int(data['Team size all employees']) - data['Employee Count']
    return data
```

```python
df = df.apply(employee_count, axis=1)
```

# Scale and PipeLine

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
X = df.drop('Dependent-Company Status', axis=1)
y = df['Dependent-Company Status']
```

```python
for i in X.columns:
    if X[i].dtype == str or X[i].dtype == object:
        label = LabelEncoder()
        X[i] = label.fit_transform(X[i])
```

```python
params = {
         'model__max_depth': [10, 20, 30, None],
         'model__max_features': ['auto', 'sqrt'],
         'model__criterion' : ["gini", "entropy", "log_loss"],
         'model__min_samples_leaf': [1, 2, 4],
         'model__min_samples_split': [2, 5, 10],
         'model__n_estimators': [60, 80, 100, 200]
}
```

```python
model = RandomForestClassifier()
scaler = StandardScaler()
```

```python
pipe = Pipeline([('scaler', scaler), ('model', model)])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
grid = GridSearchCV(pipe, param_grid=params, cv=2, verbose=2)
```

```python
grid.fit(X_train, y_train)
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```
